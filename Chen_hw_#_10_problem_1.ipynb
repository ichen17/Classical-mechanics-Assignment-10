The Lagrange of this system is

$L=\frac{1}{2}m\dot{r}^{2}+\frac{1}{2}mr^{2}\dot{\theta}^{2}+\frac{1}{2}m\dot{r}^{2}+mg\left(l-r\right)$

The corresponding momentums are

$p_{r}=2m\dot{r}$

$p_{\theta}=mr^{2}\dot{\theta}$

Therefore, after the Legrende transformation

$H=\frac{p^{2}}{4m}+\frac{p_{\theta}^{2}}{2mr^{2}}-mg\left(l-r\right)$

The equation of motion is

$\ddot{r}=\frac{C^{2}}{2r^{3}}-\frac{1}{2}g$

and

$r^{2}\dot{\theta}=C$

which $C$ is the constant.

The equilibrium happen When $\ddot{r}=0$. It means

$r_{0}=\frac{g}{\omega^{2}}$

To understand the oscillation near the position, we can assume that the object displace by $d\ll r_{0}$  Therefore, acceleration becomes

$\ddot{r}=\frac{C^{2}}{2\left(r_{0}+d\right)^{3}}-\frac{1}{2}g$

$\ddot{r}=-\frac{3C^{2}d}{2r_{0}^{2}}$

which satisfy Hook's law. It means it will oscillate with the frequency

$\omega_{o}=\sqrt{\frac{3}{2}}\frac{C}{r_{0}^{2}}$

because $C=r^{2}_{0}\omega$

$\omega_{o}=\sqrt{\frac{3}{2}}\omega$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(x, t, L):
    g=9.8
    return [x[1], (L**2)/(2*x[0]**3)-0.5*g]
```

```python
x_0 = np.array([1+0.001,0])
t=np.linspace(0.0, 10.0,10000)
y_sol = odeint(f_int, x_0, t, args=(np.sqrt(9.8),))[:,0]
plt.figure(figsize=(9,7),dpi=98)
plt.plot(t, y_sol,linewidth=1)
#plt.plot(x,analytic(x,np.pi/6),dashes=[5, 5, 5, 5],linewidth=2)
plt.grid(True)
#plt.xlabel('T ($\sqrt{gk}t$)',fontsize=10)
#plt.ylabel('X ($kx$)',fontsize=10)
plt.legend()
plt.show()
```
